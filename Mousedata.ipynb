```python
drugs = set()
for i in mm["Drug Regimen"]:
    drugs.add(i)
drugs
```

```python
drugRegimenData = {}
for drug in drugs:
    
    drugRegimenData[drug] = mm[mm["Drug Regimen"] == drug]["Mouse ID"].tolist()
```

```python
# I created a dictionary where I seperated mouse id by drug regimen, namely drugRegimenData
drugRegimenData
```

```python
newDrugDataFrame = {}
for drug in drugRegimenData.keys():
    totalTumorVolume = 0
    for mouseId in drugRegimenData[drug]:
       print(sr[sr["Mouse ID"] == mouseId]["Tumor Volume (mm3)"].mean)
    newDrugDataFrame[drug] = totalTumorVolume
```
